```python
import numpy as np

def rgb2grayscale(rgb):
    coeffs = np.array ([0.299, 0.587, 0.114])
    return np.dot(rgb, coeff)

pi = np.pi

f = 2
T = 2
a = 2
fs = 100 #sampling rate

time_array = np.linspace(0, T, fs*T)
time_array.shape
time_array

#sin wave
import matplotlib.pyplot as plt
y = a*np.sin(2*pi*f*time_array)

plt.plot(time_array, y)
plt.title('2hz sine wave')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)
```

```python
#slide_2_image_2 sampled
fs_sampled = 5
tvec_sampled = np.linspace(0, T, T*fs_sampled)
y_sampled = a*np.sin(2*pi*f*tvec_sampled)

#plot
plt.stem(tvec_sampled, y_sampled)
plt.title('sampled')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)
```

```python
#3
y_discrete = np.round(a*np.sin(2*pi*f*time_array))
plt.plot(time_array, y_discrete)
plt.title('discrete')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)

```

```python
#sampled and discretized 
plt.plot(time_array, y, 'g')
plt.stem(tvec_sampled, np.round(y_sampled))
plt.title('sampled and discretized')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)
```

```python
#subplot FINAL HW FIG 1 
f = 2
T = 2
a = 2
fs = 100 #sampling rate

time_array = np.linspace(0, T, fs*T)
time_array.shape
time_array


import matplotlib.pyplot as plt
y = a*np.sin(2*pi*f*time_array)

rows = 2
cols = 2
#sin plot
axs = plt.figure(figsize=(10,8)).subplots(rows, cols)
plt.subplot(rows,cols,1)
plt.plot(time_array, y)
plt.title('2hz sine wave')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)

#sampled
fs_sampled = 5
tvec_sampled = np.linspace(0, T, T*fs_sampled)
y_sampled = a*np.sin(2*pi*f*tvec_sampled)

#sampled plot
plt.subplot(rows,cols,2)
plt.stem(tvec_sampled, y_sampled)
plt.title('sampled')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)

#discretized
plt.subplot(rows,cols,3)
y_discrete = np.round(a*np.sin(2*pi*f*time_array))
plt.plot(time_array,y_discrete)
plt.title('discretized')
plt.xlabel('time(s)')
plt.autoscale(enable=True, axis='x', tight=True)

#sampled discretized
plt.subplot(rows,cols,4)
plt.plot(time_array,y,'g')
plt.stem(tvec_sampled, np.round(y_sampled))
plt.title('sampled and discretized')
plt.xlabel('time (s)')
plt.autoscale(enable=True, axis ='x',tight=True)
```
